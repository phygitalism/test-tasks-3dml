# Описание задачи

Необходимо подготовить приложение с демонстрацией работы модели по классификации 3D примитивов из предыдущего задания. 

Требуемая функциональность:

* Возможность загрузить свою 3D модель в формате `.obj`.
* Отображение степени уверенности модели в предсказанном классе. Например, в виде столбчатой диаграммы. 
* Визуализация загруженной 3D модели.

Для этого необходимо:
* Подготовить код и веса модели из Task3.
* Написать Dockerfile, который будет запускать ваше приложение.

Dockerfile должен содержать все необходимые шаги по установке и настройке приложения. После команды `docker run` приложение сразу должно запускаться без необходимости задания какой-либо команды.

Вы можете использовать все возможности Docker, которые доступны в версии 19.04 или 20.10.

Если ваше приложение будет состоять из нескольких частей, то можно использовать [Docker Compose](https://github.com/docker/compose) для запуска.

Предполагается, что все расчёты будут происходить на CPU. При необходимости у нас есть возможность запустить контейнер с поддержкой GPU в [nvidia-docker](https://github.com/NVIDIA/nvidia-docker).
